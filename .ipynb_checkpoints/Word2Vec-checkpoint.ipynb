```python
import gensim, logging
from string import punctuation
import numpy as np
import pandas as pd
from tqdm import tqdm
from gensim.models.doc2vec import LabeledSentence
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

Z = pd.concat([train, test], ignore_index=True)
del Z['test_id']
del Z['id']
del Z['qid1']
del Z['qid2']
Z.head()
```

```python
train_sels = ~isnan(Z.is_duplicate)
test_sels = isnan(Z.is_duplicate)

train_inds = find(train_sels)
test_inds = find(test_sels)

del train_sels, test_sels
```

```python
questions = np.concatenate([Z['question1'], Z['question2']])
questions = questions.astype('str')
```

```python
sentences = []
exclude = set(punctuation + '0123456789'+u'–—'+u'«»')

for q in tqdm(questions):
    s = q
    s = ''.join(ch for ch in s if ch not in exclude)
    sentences.append(s)
```

```python
model = gensim.models.Word2Vec(sentences, min_count=1)
```

```python
sentences = []
for i in tqdm(range(len(questions))):
    l = LabeledSentence(words = questions[i].split(), tags = ['SENT_%s' % i])
    sentences.append(l)
```

```python
model = gensim.models.doc2vec.Doc2Vec(sentences)
```

```python
second = 2750086
sims = []

for i in tqdm(test_inds):
    sim = model.n_similarity('SENT_' + str(i), 'SENT_' + str(second + i))
    sims.append(sim)
```
